```python
from itertools import combinations
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Function to compute silhouette scores
def compute_silhouette_scores(df, features, max_k):
    """
    Computes silhouette scores for KMeans clustering with different numbers of clusters.

    Parameters:
        df (pd.DataFrame): The input dataframe.
        features (list): List of feature column names to use for clustering.
        max_k (int): Maximum number of clusters to evaluate.

    Returns:
        A tuple containing the best silhouette score and the corresponding number of clusters (k).
    """
    results = []
    for k in range(3, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(df[features])
        score = silhouette_score(df[features], labels)
        results.append({'k': k, 'silhouette_score': score})
    
    results_df = pd.DataFrame(results)
    print(f"Silhouette Scores for features {features}:")
    print(results_df)

    # Find the best silhouette score and the corresponding k
    best_result = results_df.loc[results_df['silhouette_score'].idxmax()]
    best_k = best_result['k']
    best_score = best_result['silhouette_score']

    return best_score, best_k

# Function to test all combinations of 5 features
def test_feature_combinations(df, feature_list, max_k):
    """
    Tests combinations of 5 features by removing one feature at a time.

    Parameters:
        df (pd.DataFrame): The input dataframe.
        feature_list (list): List of features to combine.
        max_k (int): Maximum number of clusters to evaluate.

    Returns:
        A DataFrame with the results for each feature combination.
    """
    results = []
    
    # Generate all combinations by removing one feature at a time
    for combination in combinations(feature_list, len(feature_list) - 1):  # Always keep 5 features
        features = list(combination)
        best_score, best_k = compute_silhouette_scores(df, features, max_k)
        results.append({
            'removed_feature': list(set(feature_list) - set(features))[0],  # Feature that was removed
            'remaining_features': features,
            'best_k': best_k,
            'best_silhouette_score': best_score
        })
    
    results_df = pd.DataFrame(results)
    return results_df


max_k = 6  # Maximum number of clusters to test

results_df = test_feature_combinations(df, p1_feat, max_k)

# Display results
print("Results for removing one feature at a time:")
print(results_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def compute_silhouette_scores_range(df, features, min_k=3, max_k=7):
    """
    Computes silhouette scores for KMeans clustering with a range of cluster numbers.

    Parameters:
        df (pd.DataFrame): The input dataframe.
        features (list): List of feature column names to use for clustering.
        min_k (int): Minimum number of clusters (inclusive). Default is 3.
        max_k (int): Maximum number of clusters (inclusive). Default is 7.

    Returns:
        dict: A dictionary with the number of clusters (k) as keys and silhouette scores as values.
    """
    silhouette_scores = {}

    for k in range(min_k, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(df[features])
        score = silhouette_score(df[features], labels)
        silhouette_scores[k] = score

    return silhouette_scores
```

```python
# Compute silhouette scores for k between 3 and 7
scores = compute_silhouette_scores_range(df, p2_feat, min_k=3, max_k=7)

# Display results
print("Silhouette Scores:")
for k, score in scores.items():
    print(f"k={k}: {score}")
```
